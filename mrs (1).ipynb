```python
import ast
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
def convert3(obj):
  L=[]
  count = 0;
  for i in ast.literal_eval(obj):
    if(count < 3):
      L.append(i['name'])
      count+=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #convert string into list
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ", "") for i in x])
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres']  + movies['keywords']  + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df = movies[['id', 'title', 'tags']]
```

```python
new_df = movies[['id', 'title', 'tags']].copy()
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x).replace(",", " "))
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
import matplotlib.pyplot as plt
from collections import Counter

def plot_recommended_genres(movie_title):
    # Get recommendations
    movie_index = new_df[new_df['title'] == movie_title].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    recommended_titles = [new_df.iloc[i[0]].title for i in movies_list]

    # Get genres of recommended movies
    recommended_genres = []
    for title in recommended_titles:
        genres = movies[movies['title'] == title]['genres'].iloc[0]
        recommended_genres.extend(genres)

    # Count genres
    genre_counts = Counter(recommended_genres)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.bar(genre_counts.keys(), genre_counts.values())
    plt.title(f"Genre Distribution of Movies Recommended for '{movie_title}'")
    plt.xlabel("Genre")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Example
plot_recommended_genres('Batman Begins')
```
